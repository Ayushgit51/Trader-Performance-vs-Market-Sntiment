```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load datasets
fear_greed = pd.read_csv('fear_greed_index.csv')
trades = pd.read_csv('historical_data.csv')
```

```python
# 1. Clean Fear & Greed Data
fear_greed['date'] = pd.to_datetime(fear_greed['date'])

# 2. Clean Trading Data
# Note: 'dayfirst=True' is critical because IST format is DD-MM-YYYY
trades['Timestamp IST'] = pd.to_datetime(trades['Timestamp IST'], dayfirst=True)
trades['trade_date'] = trades['Timestamp IST'].dt.normalize()
```

```python
# 3. Handle specific data types
trades['Closed PnL'] = pd.to_numeric(trades['Closed PnL'], errors='coerce').fillna(0)

print(f"Cleaned {len(trades)} trade records and {len(fear_greed)} sentiment records.")
```

```python
# Merging the sentiment classification and numerical value into the trade log
df_merged = pd.merge(
    trades, 
    fear_greed[['date', 'value', 'classification']], 
    left_on='trade_date', 
    right_on='date', 
    how='left'
)
```

```python
# Check for alignment issues (trades without matching sentiment)
missing_sentiment = df_merged['classification'].isnull().sum()
if missing_sentiment > 0:
    print(f"Warning: {missing_sentiment} trades did not have a matching sentiment date.")
    df_merged['classification'] = df_merged['classification'].fillna('Neutral') # Defaulting for continuity
```

```python
# Filter for trades where PnL was realized
closed_trades = df_merged[df_merged['Closed PnL'] != 0].copy()

# Metric Calculations
sentiment_analysis = closed_trades.groupby('classification').agg({
    'Closed PnL': ['count', 'sum', 'mean', 'std'],
    'value': 'mean'
}).reset_index()

sentiment_analysis.columns = ['Sentiment', 'Trade Count', 'Total PnL', 'Avg PnL', 'PnL Volatility', 'Avg Index Value']

# Calculate Win Rate
win_rates = closed_trades.groupby('classification')['Closed PnL'].apply(lambda x: (x > 0).sum() / len(x) * 100)
sentiment_analysis['Win Rate %'] = sentiment_analysis['Sentiment'].map(win_rates)

display(sentiment_analysis.sort_values('Total PnL', ascending=False))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# 1. Aggregate Daily behavior features
daily_data = df_merged.groupby('trade_date').agg({
    'Closed PnL': ['sum', 'std'],            # Performance & Volatility
    'Trade ID': 'count',                     # Activity level
    'Size USD': 'mean',                      # Risk appetite
    'value': 'first'                         # Sentiment Index (today's value)
})


# Flatten column names
daily_data.columns = ['daily_pnl', 'pnl_volatility', 'trade_count', 'avg_size', 'sentiment_value']
daily_data = daily_data.fillna(0).reset_index()
```

```python
# 2. Add Behavioral Momentum (Rolling 3-day PnL)
daily_data['pnl_3d_rolling'] = daily_data['daily_pnl'].rolling(window=3).mean()
```

```python
# 3. CREATE THE TARGET (Predicting Tomorrow)
# We shift the PnL by -1 so that today's features are aligned with tomorrow's outcome
daily_data['tomorrow_pnl'] = daily_data['daily_pnl'].shift(-1)

# Profitability Bucket: 1 if profitable tomorrow, 0 if loss/break-even
daily_data['target_profit'] = (daily_data['tomorrow_pnl'] > 0).astype(int)

# Drop the last row because we don't know tomorrow's result for the most recent date
df_ml = daily_data.dropna().copy()

print(f"Dataset ready for modeling. Total days: {len(df_ml)}")
print(f"Target Distribution: {df_ml['target_profit'].value_counts(normalize=True)}")
```

```python
# Select features that represent 'Sentiment + Behavior'
feature_cols = ['sentiment_value', 'trade_count', 'avg_size', 'daily_pnl', 'pnl_3d_rolling', 'pnl_volatility']
X = df_ml[feature_cols]
y = df_ml['target_profit']

# Split data: Use the first 80% to train and last 20% to test (chronological)
split_idx = int(len(df_ml) * 0.8)
X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]
```

```python
# Train the model
rf_model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
rf_model.fit(X_train, y_train)

# Evaluate
y_pred = rf_model.predict(X_test)
print(f"Model Accuracy: {accuracy_score(y_test, y_pred):.2%}")
print("\nDetailed Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Visualize Feature Importance
importances = pd.DataFrame({
    'Feature': feature_cols,
    'Importance': rf_model.feature_importances_
}).sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x='Importance', y='Feature', data=importances, palette='magma')
plt.title('Predictive Power: What Influences Tomorrow\'s Profit?')
plt.show()

# Generate Automated Insight
top_f = importances.iloc[0]['Feature']
print(f"--- Actionable Conclusion ---")
print(f"The most predictive factor for your next-day success is: {top_f.upper()}.")
print(f"Insight: This suggests that your {'market context' if 'sentiment' in top_f else 'own recent behavior'} ")
print(f"is the primary driver of whether you will have a winning or losing day tomorrow.")
```
